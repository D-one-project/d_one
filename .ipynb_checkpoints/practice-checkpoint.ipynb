```python
import pandas as pd
import numpy as np
```

### 1. Create 5 random variables range from 200~1200 with 50 data points called a-e.

```python
cols = ['a','b','c','d','e']
df = pd.DataFrame()

for c in cols:
    df[c] = np.random.uniform(200,1200,50)
```

```python
df
```

### 2. Create one variable range from 1-5 called group1

```python
data1 = [1,2,3,4,5]

rpt = int(len(df)/len(data1))
mod = len(df)%len(data1)

df2 = pd.DataFrame({'group1':data1 * rpt + data1[0:mod]})
```

```python
df2.head(7)
```

### 3. Create one variable range from a-c called group2

```python
data2 = ['a','b','c']

rpt = int(len(df)/len(data2))
mod = len(df)%len(data2)

df3 = pd.DataFrame({'group2':data2 * rpt + data2[0:mod]})

```

```python
df3.head(5)
```

### 4. Use group by with sum, mean, median, sd, and get the first value or last value from the index to aggregate

#### - merged dataframe

```python
merged_df = pd.concat([df, df2, df3], axis=1)

merged_df
```

#### - describe()

```python
merged_df.describe()
```

### 4. Use group by with sum, mean, median, sd, and get the first value or last value from the index to aggregate

```python
merged_df.groupby('group1').mean(numeric_only=True)
```

```python
merged_df.groupby('group1').median(numeric_only=True)
```

```python
merged_df.groupby('group1').std(numeric_only=True)
```

### 5. Use pivot_table to replicate the result from 4

```python
# In pivot_table, index is like.. Y axis? and columns arg are like X axis? 
# columns arg to group column titles
```

```python
# To play around with certain rows of a DataFrame, iloc first to filter and put it into pivot_table?
```

```python
pvtbl = pd.pivot_table(merged_df.iloc[1:10], index=['group2'], aggfunc=np.sum)

pvtbl
```

```python
pd.pivot_table(merged_df.iloc[1:10], index=['group2','group1'], aggfunc=np.sum)
```

```python
pd.pivot_table(merged_df.iloc[1:10], index=['group2'], columns=['group1'], values=['a','b'], fill_value='fill_test' ,aggfunc=np.sum)
```

### 6. Use melt to have a data set as follows from 5
- Field, value, type

```python
# Melt is like.. data re-arrange tool (reshaping data, restructuring data, ... ** unstructured raw data can be reshaped by using melt())?
```

```python
pvtbl
```

```python
mtdf = pd.melt(pvtbl, value_vars=['a','b','c','d','e'], value_name='random_value', var_name='group2')

mtdf
```

```python
list = []

for value in mtdf['random_value']:
    list.append(value.__class__)
```

```python
list
```

```python
mtdf['type'] = list
```

```python
mtdf
```

### 7. Utilize apply lambda for free calculation

```python
lmdf = mtdf
```

```python
# lambda is like a function
```

```python
lmdf100 = lmdf.apply(lambda x : x / 100 if x.name in ['random_value'] else x, axis=0)

lmdf100
```
